```python
import numpy as np
```

```python
f = open('2.RAW', 'rb')
contents = f.read()
f.close()
```

```python
from PIL import Image

image = np.array([b for b in contents])
m,n = 240, 320
matrix = image.reshape((480,640))
first, second, third, fourth = np.zeros(m*n), np.zeros(m*n), np.zeros(m*n), np.zeros(m*n)
index = 0



for j in range(m):    #rows
    for i in range(n):    #columns
        first[index] = matrix[j*2,i*2]     #even row even column
        second[index] = matrix[j*2+1, i*2+1]     #odd row odd column
        third[index] = matrix[j*2+1, i*2]        #even row odd column
        fourth[index] = matrix[j*2, i*2+1]       #odd row even column
        
        index += 1
        
temp = np.concatenate((first, second, third, fourth), axis=None)

image = np.array(temp)
matrix = image.reshape((480,640))

img = Image.fromarray(np.uint8(matrix), 'L')
img.show()
```

```python
newh, neww = (int) (h/2), (int) (w/2)
new_mat = np.zeros((newh, neww))

# get every fourth pixel
for i in range(0, newh):
    for j in range(0, neww):
        new_mat[i,j] = matrix[i*2, j*2]
        
img = Image.fromarray(np.uint8(new_mat), 'L')
img.show()
```

```python
img.save('4comp.png')
```

### Put 2 values into 1 byte (8-bit --> 4-bit)

```python
# compress each pair of 8-bit pixels into a single byte
d2by4 = (image[::2] & 0xf0) | (image[1::2] >> 4)
len(d2by4)
```

```python
#unpack on the other side back into 8-bit values
f = open('test.bin', 'rb')
contents = f.read()
d2by4 = [b for b in contents]

ext_image = []

for i in range(len(d2by4)):
    high_nib = d2by4[i] & 0xf0
    low_nib = (d2by4[i] & 0x0f) << 4
    
    ext_image.append(high_nib)
    ext_image.append(low_nib)

len(ext_image)

print(ext_image)
```
